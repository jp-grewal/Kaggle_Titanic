```python
# Converting male, female values to 0, 1.
train['Sex'] = pd.get_dummies(train['Sex'])
test['Sex'] = pd.get_dummies(test['Sex'])

# Mapping different embarking features.
train['Embarked'] = train['Embarked'].map({'S': 1, 'C': 2, 'Q': 3})
test['Embarked'] = test['Embarked'].map({'S': 1, 'C': 2, 'Q': 3})
# train['Embarked_S'] = train['Embarked'].map({'S': 1, 'C': 0, 'Q': 0})
# train['Embarked_C'] = train['Embarked'].map({'S': 0, 'C': 1, 'Q': 0})
# train['Embarked_Q'] = train['Embarked'].map({'S': 0, 'C': 0, 'Q': 1})
# test['Embarked_S'] = test['Embarked'].map({'S': 1, 'C': 0, 'Q': 0})
# test['Embarked_C'] = test['Embarked'].map({'S': 0, 'C': 1, 'Q': 0})
# test['Embarked_Q'] = test['Embarked'].map({'S': 0, 'C': 0, 'Q': 1})

# # Replacing NaN values with average values.
# train['Age'] = train['Age'].fillna(train['Age'].mean())
# # dev['Age'] = dev['Age'].fillna(dev['Age'].mean())
# test['Age'] = test['Age'].fillna(test['Age'].mean())

# train['Fare'] = train['Fare'].fillna(train['Fare'].mean())
# # dev['Fare'] = dev['Fare'].fillna(dev['Fare'].mean())
# test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
```

```python
# Taken from Kaggle User 'nmei-42'.

def naive_process_title(train_test_lst):
    new = []
    for dataset in train_test_lst:
        dataset = dataset.copy(deep=True)
        titles = [name.split(',')[1].lstrip(' ').split('.')[0] for name in dataset['Name']]
        dataset = dataset.assign(Title=pd.Series(titles).values)
        new.append(dataset)
    return (new)

train, test = naive_process_title([train, test])
```

```python
# Encoding various different features to integer values

LE = preprocessing.LabelEncoder()
train['Title'] = LE.fit_transform(train['Title'])
test['Title'] = LE.fit_transform(test['Title'])

train['Ticket'] = LE.fit_transform(train['Ticket'])
test['Ticket'] = LE.fit_transform(test['Ticket'])

train['Cabin'] = train['Cabin'].fillna('JAI')
test['Cabin'] = test['Cabin'].fillna('JAI')

train['Cabin'] = LE.fit_transform(train['Cabin'])
test['Cabin'] = LE.fit_transform(test['Cabin'])
```
